## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import io
```

## Read Image 

```python
img = cv2.imread("House.jpg")
plt.imshow(img)
```

## Reading in RGB Format 

```python
img1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

## Converting to Grayscale

```python
grey_img = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
plt.imshow(grey_img)
```

## Inverting the Image

```python
inverted_img = 255-grey_img
plt.imshow(inverted_img)
```

## Generating the Blurred Image

```python
blurred_img = cv2.GaussianBlur(inverted_img,(21,21),0)
plt.imshow(blurred_img)
```

## Inverting the Blurred Image

```python
inverted_blur_img = 255-blurred_img
plt.imshow(inverted_blur_img)
```

## Converting to Pencil Sketch

```python
pencil_sketch = cv2.divide(grey_img,inverted_blur_img,scale=256)
```

```python
plt.imshow(pencil_sketch)
```
